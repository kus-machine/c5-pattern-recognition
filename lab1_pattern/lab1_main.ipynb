```python
import importlib
from cv2 import imread, imwrite, imshow, waitKey, COLOR_BGR2RGB, cvtColor, vconcat, hconcat, putText, \
    FONT_HERSHEY_SIMPLEX, resize
import time
import lab1
from os import mkdir, path
import numpy as np
# importlib.reload(lab1)
```

```python
filename = "field1.jpg"

scale = [512, 512]
n_iter = 1000
eps = 0.1
```

```python
img1 = imread(filename)
# scaling image to lower resolution for good performance
if scale is not None:
    img1 = resize(img1, (scale[0], scale[1]))

start_time = time.time()
img_for_sampl, LABELS = lab1.sampler(img1, eps=eps, n_iter=n_iter, h=.15, w=.44, offset=0.05)
alg_time = time.time() - start_time
print("--- %s seconds ---" % alg_time)
print("time per iter: ", alg_time / n_iter)

last_result = lab1.write_result_to_path(filename, LABELS, img_for_sampl, a=30, b=100)
lab1.show_image(last_result)
```
